## DATA VISUALIZATION

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Rel Hum_%'], bins=20, color='lightblue', kde=True)
plt.xlabel('Humidity')
plt.ylabel('Frequency')
plt.title('Humidity Distribution')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Temp_C', y='Dew Point Temp_C', color='DarkBlue')
plt.xlabel('Temperature in Celcius')
plt.ylabel('Dew Point Temperature in Celcius')
plt.title('Temperature Distribution in Celcius')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Wind Speed_km/h'], color='lightblue')
plt.xlabel('Wind Speed_km/h')
plt.title('Wind Speed Distribution')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Visibility_km'], bins=20, color='DarkBlue')
plt.xlabel('Visibility_km')
plt.ylabel('Frequency')
plt.title('Visibility_km Distribution')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Press_kPa'], bins=20, color='LightBlue', kde = 'true')
plt.xlabel('Pressure in kPa')
plt.ylabel('Frequency')
plt.title('Pressure Distribution')
plt.tight_layout()
plt.show()
```

```python
df['Date/Time'] = pd.to_datetime(df['Date/Time'])

df['Year'] = df['Date/Time'].dt.year
df['Month'] = df['Date/Time'].dt.month
df['Day'] = df['Date/Time'].dt.day
df['Hour'] = df['Date/Time'].dt.hour

plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='Temp_C', data=df)

plt.title('Temperature Over Months')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date/Time', y='Wind Speed_km/h', data=df)
plt.title('Wind Speed Over Time')
plt.xlabel('Date/Time')
plt.ylabel('Wind Speed (km/h)')
plt.show()
```
